```python
"""
-Le nombre de partitions doit etre au minimum le nombre de workers
-Les partitions definissent la taille des listes sur lesquels les workers travaillent
-Travailler de plus petites partitions (donc + de partitions) peut permettre plus de parall√©lisation
"""
# nombre de partition (cloisons) par default dans le SC = Le nombre de Workers
from pyspark import SparkContext
# spark context avec 4 workers
sc = SparkContext(master="local[4]")
```

```python
A = sc.parallelize(range(1000000))
# nombre de partitions : 4 car 4 wrkr
print(A.getNumPartitions())
```

```python
# changer nb de partitions
D=A.repartition(10)
print(D.getNumPartitions())
```

```python
# creer un RDD avec nos partitions custom
A = sc.parallelize(range(1000000),numSlices = 10)
print(A.getNumPartitions())
```

```python
# 10 partitions de 100k elem :
A = sc.parallelize(range(1000000),numSlices = 10)
A.map(lambda x:x*x)
print(A.glom().map(lambda x:len(x)).collect())
```

```python
# filtrer 10% des partitions

B=A.filter(lambda pair: pair[0]%5==0)
print(B.glom().map(len).collect())
```

```python
# partitioner
C=B.map(lambda p:(p[1]/10,pair[1])).partitionBy(10)
print(C.glom().map(len).collect())
```

```python
C=B.repartition(10)
print(C.glom().map(len).collect())
```
